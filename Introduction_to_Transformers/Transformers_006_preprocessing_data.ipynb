## 3 Preprocessing  
In this phase we are preprocessing the downloaded dataset

```python
from transformers import AutoTokenizer
```

```python
ckpt = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(ckpt)
```

```python
tokenized_sentence_1 = tokenizer(raw_dataset["train"]["sentence1"])
tokenized_sentence_2 = tokenizer(raw_dataset["train"]["sentence2"])
```

```python
tokenizer("Sentence one", "sentence two")
```

```python
def tokenizer_func(example):
  return tokenizer(
    example["sentence1"],
    example["sentence2"],
    truncation=True)
```

```python
tokenized_dataset = raw_dataset.map(tokenizer_func, batched=True)
tokenized_dataset
```

for every sentence in our dataset we have done the preprocessing, i.e. we have obtained features: ['sentence1', 'sentence2', 'label', 'idx', 'input_ids', 'token_type_ids', 'attention_mask'] for all our data

## 4 Dynamic Padding

```python
from transformers import DataCollatorWithPadding
#this will help with data padding part
```

```python
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

## 5 Training
using Keras API

```python
from transformers import TrainingArguments

training_args = TrainingArguments("test-trainer") # test trainer is a default trainer available and it contains configuration
training_args
```

```python
ckpt
```

```python
# creating model

from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(ckpt, num_labels=2)
```

```python
from transformers import Trainer

trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["validation"],
    data_collator=data_collator, # for dynamic padding
    tokenizer=tokenizer
)
```

```python
trainer.train()
```

```python
predictions = trainer.predict(tokenized_dataset["validation"])
predictions
```
