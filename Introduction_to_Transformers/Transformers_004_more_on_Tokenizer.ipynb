## handling multiple sequences

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
ckpt = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(ckpt)
model = AutoModelForSequenceClassification.from_pretrained(ckpt)
```

```python
example = "I've been exploring Transformers for Natural Language Processing"
```

```python
tokens = tokenizer.tokenize(example)  # tokenizing the sentence, i.e. breaking into word tokens
ids = tokenizer.convert_tokens_to_ids(tokens) # converting word tokens into ids
input_ids = torch.tensor(ids) # coverting ids as a tensor
```

```python
tokens
```

```python
ids
```

```python
input_ids
```

```python
model(input_ids)
```

- for : **model(input_ids)**  we are trying to supply tensor_id directly to model.
- we have only got a one dimesional tensor, we will need a 2D tensor to pass the information.
- to convert **input_ids** to 2D we use torch.unsqueeze

```python
torch.unsqueeze(input_ids, dim=0)
```

```python
model(torch.unsqueeze(input_ids, dim=0))
```

#### Aternately we can give "return_tensors = 'pt'"parameter in the tokenizer itself to get the pytorch tensors


```python
tokenized_inputs = tokenizer(example, return_tensors='pt')
```

```python
tokenized_inputs["input_ids"]
```

```python
model(tokenized_inputs["input_ids"])
```

notice that tensor outputs will be diffrent if we use unsqueeze method  and parmaeter in tokenizer method

#### Padding

```python
tokenizer.pad_token_id
```

```python
# padding variable length sentences
example_1 = "I've been exploring Transformers for Natural Language Processing."
example_2 = "I've been exploring Transformers."
```

```python
token_1 = tokenizer.tokenize(example_1)
print(token_1)
token_2 = tokenizer.tokenize(example_2)
print(token_2)
```

```python
id_1 = tokenizer.convert_tokens_to_ids(token_1)
id_2 = tokenizer.convert_tokens_to_ids(token_2)
```

```python
len(id_1)
```

```python
len(id_2)
```

```python
batch_id = [
            id_1,
            id_2
]
batch_id
```

```python
model(torch.tensor([id_1])).logits
```

```python
model(torch.tensor([id_2])).logits
```

```python
# pad sequence length must be same
model(torch.tensor(batch_id)).logits
```

```python
batch_id_same_seq_len = [
            id_1,
            id_1
]
```

```python
model(torch.tensor(batch_id_same_seq_len)).logits
```

```python
# pad max length to 11

max_len = 11

while  len(id_2)!= max_len:
  id_2.append(tokenizer.pad_token_id)
```

```python
len(id_2)
```

```python
id_2
```

```python
batch_id = [
            id_1,
            id_2
]
batch_id
```

```python
model(torch.tensor(batch_id)).logits
```
