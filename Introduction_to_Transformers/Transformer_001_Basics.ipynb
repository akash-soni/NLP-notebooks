## Zero shot classifire

Zero shot means seeing the data for first time it is able to do the prediction task

```python
classifier_zero_shot = pipeline("zero-shot-classification")
```

### MNIL means it Multi Genere Natural Language Inference

This means that given a sentence and candidate_labels, it can do a hypothesis analysis and predict the probabilities of given label and choose the label that have highest probability

```python
classifier_zero_shot(
    "This course is about artificial intelligence, Here we will learn theory and practical both.",
    candidate_labels=["politics", "business", "education", "workshop", "games"],
)
```

```python
classifier_zero_shot(
    "Narendra Modi is PM of India.",
    candidate_labels=["politics", "business", "education", "workshop", "games"],
)
```

```python
classifier_zero_shot(
    "Assassins creed has a character Ezio ",
    candidate_labels=["politics", "business", "education", "workshop", "games"],
)
```

## Text Generation

```python
generator = pipeline("text-generation")
```

```python
generator("This course is about artificial intelligence")[0]["generated_text"]
```

```python
# supplying model alongwith text-generation
# the advantage is we can supply model according to our compute requirements
generator = pipeline("text-generation", model="distilgpt2")
```

```python
generator("This course is about artificial intelligence",
          max_length=40,
          num_return_sequences=3)
```

```python
generator("The movie was horrible",
          max_length=40,
          num_return_sequences=2)
```

```python
generator("<p> This is HTML </p>",
          max_length=40,
          num_return_sequences=2)
```

```python
generator("SELECT * FROM TABLE",
          max_length=40,
          num_return_sequences=2)
```

for the above HTML and SQL cases it is not doing very good, for these tasks we can go for FINE-TUNING

## Masking filling: Fill in the blanks types

```python
unamasking = pipeline("fill-mask")
```

```python
unamasking("This course is about <mask> intelligence", top_k=3)
```

```python
unamasking("I <mask> you", top_k=3)
```

## Named entity Recognition

## Question answering

```python
QA = pipeline("question-answering")
```

```python
QA(
    question="Wher do I work?",
    context="I am Sunny and I work in iNeuron Intelligence. I live in Bangalore."
)
```

```python
QA(
    question="How far is India from Srilanka?",
    context="India’s frontier, which is roughly one-third coastline, abuts six countries. It is bounded to the northwest by Pakistan, to the north by Nepal, China, and Bhutan; and to the east by Myanmar (Burma). Bangladesh to the east is surrounded by India to the north, east, and west. The island country of Sri Lanka is situated some 40 miles (65 km) off the southeast coast of India across the Palk Strait and Gulf of Mannar."
)
```

## Text Summarization

```python
summarization = pipeline("summarization")
```

Abstraction: It generates new text to summarize
Extraction: It extract new text to summarize

```python
in_text = """Gradio is an open-source Python library that is used to build machine learning and data science demos and web applications.
With Gradio, you can quickly create a beautiful user interface around your machine learning models or data science workflow and let people "try it out" by dragging-and-dropping in their own images, pasting text, recording their own voice, and interacting with your demo, all through the browser."""

print(in_text)
summarization(in_text, max_length=56)[0]["summary_text"]
```
